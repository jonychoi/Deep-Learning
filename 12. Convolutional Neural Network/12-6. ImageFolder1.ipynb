Here, we are going to learn about the ```ImageFolder``` of PyTorch's ```torchvision.datasets```, that allows us to add as a datasets with our custom data.

## Imports

````python
import torchvision
from torchvision import transforms
````

````python
from matplotlib.pyplot import imshow
%matplotlib inline
````

## Define torchvision.transforms propertise to create the datasets

````python
trans = transforms.Compose([
    transforms.Resize((64, 128))
])

train_data = torchvision.datasets.ImageFolder(root = 'catsanddogs/origin_data/', transform = trans)
````

## Save the transformed images from origin_data to train_data

````python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)

    if(label == 0):
        data.save('catsanddogs/train_data/cats/{}_{}.jpeg'.format(label, num))
    else:
        data.save('catsanddogs/train_data/dogs/{}_{}.jpeg'.format(label, num))
````
